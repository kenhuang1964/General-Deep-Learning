## Sentiment Analysis

```python
from transformers import pipeline 

classifier = pipeline("sentiment-analysis")
classifier("Deep learning is going to change the future! I am so glad that I am learning it right now")
```

```python
classifier(["Man coding is too fun to be true lol",
            "Are you coding for fame or for fun?",
            "If you are coding for popularity or fame then you should stop",
            "The fame you feel will soon dissipate as you advance further"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "Do not learn deep learning or machine learning if you haven't learned the math",
    candidate_labels=["advice", "business", "politics", "data science"]
)
```

```python
generator = pipeline("text-generation")
generator("Learning the math required for machine learning and deep learning will allow you to ")
```

```python
generator = pipeline("text-generation")
generator(
    "Math and problem solving are absolutely essential to succeed in DL and ML",
    max_length = 50,
    num_return_sequences=2,
)
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question="What is your approach to learning deep learning and machine learning effectively?",
    context="I am trying to learn deep learning and machine learning as a student",
)
```
